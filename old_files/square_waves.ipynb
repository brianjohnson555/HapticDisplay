```python
import numpy as np
import matplotlib.pyplot as plt

def squarewave(t, frequency=1, amplitude=1, offset=0, phase=0):
    y = np.sin(2*frequency*np.pi*t-phase*(np.pi))
    y[y>0]=amplitude
    y[y<=0]=-amplitude
    y = y+offset
    return y

t = np.linspace(0,4,401)
HV = squarewave(t, 2, 1, 0, 0)
GND = squarewave(t, 2, 1, 0, 1)

fig, axs = plt.subplots(3)
axs[0].plot(t, HV)
axs[1].plot(t, GND)
axs[2].plot(t, HV-GND)
```
